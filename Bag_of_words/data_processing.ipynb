```python
IN_MASTER = r'/data/workspace_files/Loughran-McDonald_MasterDictionary_1993-2021.csv'
folder_path = r'/data/workspace_files/doc_dictionaries'
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
```

```python
master_dictionary = md.load_masterdictionary(IN_MASTER, print_flag=True, get_other=False)
lookup = rd.create_lookup_dictionary(master_dictionary)
```

```python
dic = pd.read_csv(IN_MASTER)
len(dic)
```

```python
def process_file(files, target_year):
    folder_path = '/data/workspace_files/doc_dictionaries'
    file_num = 1
    for curr_file in files:
        file_path = os.path.join(folder_path, curr_file)
        #print('Starting for', curr_file)
        
        with open(file_path) as f_in:
            for count, line in enumerate(f_in):
                try:
                    header, docdict = rd.read_docdict(line, lookup)
                    if header.form_type not in ['10-K', '10-Q', '10-K405', '10-KSB'] or datetime.strptime(str(header.conformed_period_report), '%Y%m%d').date().year != target_year:
                        continue
                    
                    row = {
                        'cik': header.cik,
                        'filing_date': datetime.strptime(str(header.filing_date), '%Y%m%d').date(),
                        'accession_number': header.accession_number,
                        'conformed_period_report': datetime.strptime(str(header.conformed_period_report), '%Y%m%d').date(),
                        'form_type': header.form_type,
                        'company_name': header.company_name,
                        'total_words': header.total_words,
                        'doc_dict_text': str(docdict)  # Convert doc_dict to text
                    }
                    yield row  # Yield each row instead of appending to a list
                except Exception as e:
                    print(f'some issue with {header.cik} {header.filing_date} {header.conformed_period_report}: {e}')
        
        # Free up memory
        gc.collect()
        print(f'Finished file {file_num}: {curr_file}')
        file_num += 1

# Usage example
for year in range(2000, 2018):  # Example range of years
    rows = process_file(files, year)
    df = pd.DataFrame(rows)
    parquet_file_path = os.path.join('/data/workspace_files/parquet_files', f'file_{year}.parquet')
    df.to_parquet(parquet_file_path, index=False)
    print(f'Processed year {year}')
```
